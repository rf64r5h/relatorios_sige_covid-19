# Número de casos por faixa etária

## Gŕafico de barras

```python
# Importação das bibliotecas necessárias

import pandas as pd # Pandas é uma biblioteca de código aberto que fornece estruturas de dados de alto desempenho e fáceis de usar e ferramentas de análise de dados para a linguagem de programação Python.
import matplotlib.pyplot as plt # Matplotlib é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy.
```

```python
# Leitura dos data frames/csv
df1 = pd.read_csv("/part-00000-af9dbba4-9718-4b2b-bbfe-ef4024e867b5.c000(1).csv", delimiter=";", quotechar='"', low_memory=False)
df2 = pd.read_csv("/part-00000-af9dbba4-9718-4b2b-bbfe-ef4024e867b5.c000.csv", delimiter=";", quotechar='"', low_memory=False)

# Concatenação dos dois arquivos CSV
dataCovid = pd.concat([df1, df2])
```

```python
dataCovid
```

```python
idade_counts = dataCovid['idade'].value_counts() # Contar a ocorrência de cada valor único na coluna 'idade' do nosso csv, value_counts() retorna uma série devalores únicos e a contagem de ocorrências
idade_counts = idade_counts.sort_index() # Classificando por ordem crescente

# Criar o gráfico de barras
plt.bar(idade_counts.index, idade_counts.values)

plt.xlabel("Idade") # Label/titulo do eixo x
plt.ylabel("Número de casos") # Label/titulo do eixo y
plt.title("Casos de COVID-19 por Idade") # Titulo do gráfico

plt.show() # Exibir o gráfico de barras
```

# Distribuição de casos por sexo

## Gráfico de pizza

```python
sexo_counts = dataCovid['sexo'].value_counts() # Contagem dos valores únicos(index) e a contagem de sua ocorrência

#Grafico de pizza da distribuição de casos por sexo
plt.pie(sexo_counts.values, labels=sexo_counts.index, autopct='%1.1f%%') # Criar o gráfico de pizza, O parâmetro autopct='%1.1f%%' formata a exibição dos percentuais das fatias.
plt.title("Distribuição de casos por sexo") # Titulo do gráfico

plt.show() # Exibir o gráfico de pizza
```

# Top N sintomas mais comuns

## Gráfico de barra horizontal

```python
sintomas_counts = dataCovid['sintomas'].value_counts() # Contagem dos index e valores

#Selecionar os N sintomas mais comuns
n = 10  #Defina o valor de N

top_n_sintomas = sintomas_counts.head(n)

#Gráfico de barra horizontal
plt.barh(top_n_sintomas.index, top_n_sintomas.values)

#Configuração dos rotulos e titulo do gráfico
plt.xlabel("Número de Casos")
plt.ylabel("Sintomas")
plt.title(f'Top {n} Sintomas Mais Comuns')

#Exibir o gráfico
plt.show()


```

# Tendência dos casos ao longo das diferentes faixas etárias

```python
casos_por_idade = dataCovid.groupby('idade').size() # Agrupar os casos por faixa etária e contar a ocorrência
casos_por_idade = casos_por_idade.sort_index() # Ordenar as faixas etárias de forma crescente

plt.plot(casos_por_idade.index, casos_por_idade.values, marker='.') # Criar o gráfico de linha, marker='o' (qual caracter para marcar os pontos da linha)

# Configurar os rótulos e título do gráfico
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Casos')
plt.title('Tendência dos Casos ao Longo das Diferentes Faixas Etárias')

plt.show() # Exibir o gráfico
```

# Distribuição percentual de casos entre profissionais de saúde e não profissionais de saúde,

```python
# Contagem de ocorrência de casos para profissionais de saúde e não profissionais de saúde
profissional_saude_counts = dataCovid['profissionalSaude'].value_counts()

# Criar o gráfico de pizza
plt.pie(profissional_saude_counts, labels=profissional_saude_counts.index, autopct='%1.1f%%')

# Configurar o título do gráfico
plt.title('Distribuição Percentual de Casos entre Profissionais de Saúde e Não Profissionais de Saúde')

# Exibir o gráfico
plt.show()
```
